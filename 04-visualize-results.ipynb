```python
A_dic = nx.read_gpickle('03-A_exp/04122020-A_all_exp.json')
```

```python
dd, pdd, sorted_pdd = f.compute_dd_probd(A_dic)
k0_list = list(dd['p005'].keys())
```

```python
fig= plt.figure(figsize=(15, 4))
i=0
n=300
for k in [0,9]:
    plt.subplot(1, 2, i+1)
    i += 1
    plt.scatter(int(k0_list[k]), 1, label='p = 0') #regular lattice
    plt.plot(f.supports(sorted_pdd['p005'], k), f.prob_distrib(sorted_pdd['p005'], k), label='p = 0.05')
    plt.plot(f.supports(sorted_pdd['p01'], k), f.prob_distrib(sorted_pdd['p01'], k), label= 'p = 0.1')
    plt.plot(f.supports(sorted_pdd['p05'], k), f.prob_distrib(sorted_pdd['p05'], k), label= 'p = 0.5')
    plt.plot(f.supports(sorted_pdd['p09'], k), f.prob_distrib(sorted_pdd['p09'], k), label= 'p = 0.9')
    plt.legend()
    plt.title('Degree Distribution of reference models with fixed k =' + str(k0_list[k]))
    plt.savefig('degreedistribk10')
    plt.xlabel('k')
    plt.ylabel('p(k)')
```
